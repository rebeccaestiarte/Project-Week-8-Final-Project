#### Main Hipotesis

My hipotesis is that the practice of Mindful Meditation affects the brainwaves activity even for inexperienced meditators. I also want to prove that the change in brainwaved frequency pattern experienced in meditation is beneficial for the practicioner.

#### The Experiment:

###### Instruments:
Using Muse wearable brain sensing headband, manufactured by InteraXon, I was able to measure the brain activity of a sample group via 4 electroencephalography (EEG) sensors. As Interaxon has retired from the market the Software Developer Kit (SDK) for Muse, I used a third party app named "Mind Monitor", by James Clutterbuck. The app is able to process the EEG data and output a single combined average value for the five absolute brain wave values. It can output the data into Excel compatible CSV (Comma Separated Values) files that I directly uploaded into a Google Drive folder from my smartphone.

###### Description:
The experiment consisted in getting the participants, once at the time, in a isolated room and lasted aproximately 20 minutes per participant. 

In the first part of the study, I placed the headband on the participant and started the recording with the application. During the first 10 minutes of the recording, I explaned the experiment to the participant to get the control data (which will be referred in this paper as "Normal State session"). I saved the data as CSV file and uploaded it on the cloud with the filename format "museMonitor -- participant name -- Normal_data--timestamp_ID.csv".

In the second part of the study, I made the participant close their eyes and listed to a 10 minutes guided Mindful meditation video https://www.youtube.com/watch?v=6p_yaNFSYao&t=461s. During this time, I recorded their brainwaves activities with the headband and uploaded the resulting csv as "museMonitor -- participant name -- Meditative_data--timestamp_ID.csv".

Finally, I asked feedback to the participant and noted down my observations.

##### Statistical Analysis

By aggregrating average brain frequencies by timestamp of participants during meditation and in control, I was able to study the different patterns in brainwaves behaviours between the two states. I drowned two main conclusions, which I proved through hypotesis testing:

1. The average brain frequencies during meditation are closer to optimal range (0) than at normal state.
2. Alpha Waves during meditation sessions predominates (have higher average values) over the other waves, which is not the case during normal state sessions.
3. While Gamma Waves seams to have more peaks during meditation, a difference cannot be proven with statistical significance.

##### Results and Conclusions

1. Studies demostrate that when brainwaves frequencies move withing the optimal range, the subject experiences a set of benefits [Lucid 2016, Kruizinga](#cite-PER-GRA:2007). The benefits related to each wave optimal frequency are: 

    - Alpha: Ideal relaxation and concentration,
    - Beta: Consistent focus, strong memory recall, high problem solving ability,
    - Delta: Healthy immune system,  restorative REM sleep,
    - Gamma: Information processing, cognition, learning, binding of senses, 
    - Theta: Maximum creativity, deep emotional connection with oneself and others, greater intuition, relaxation.

Proven that the average brain frequencies during meditation are closer to optimal range (0), we can infer that the subject is experiencing the above mentioned benefits during the meditation practice. 

2. It has been proved that mindfulness training and meditation tend to produce noticeably more alpha waves. When alpha oscillations are prominent, your sensory inputs tend to be minimized and your mind is generally clear of unwanted thoughts. Studies have shown correlation between increasing alpha brain waves and the stimulation of creativity* and minimization of depression*. Assuming the validity of those studies and my results, we can conclude that meditation increase abilty to concentrate, the overall wellbeing and stimulates creative thoughts.  


3. The fact that I couldn't prove a significant difference of Gamma waves during meditation vs control, alignes with the finding of studies about gamma waves behaviour. People with very high levels of gamma activity are exceptionally intelligent, compassionate, and have strong self-control. The only subjects that have been proven to work on signicantly high gamma activity, which relates to peak intellectual performance,  are long-term Buddhist practitioners which underwent mental training for 10,000 to 50,000 h over time periods ranging from 15 to 40 years. It is to expect that for amateur meditator the differences in gamma are non significant*.

##### Machine Learning Model

The machine learning model algorithm aims to identify if a person is in state on meditation or not based on the brainfrequencies the headband registers. It could be used to give real time feedback to the user during a meditation session by alerting the user when its mind is wondering. The first step was processing the data to train the model: firstly, I divided each recording session into four quartiles to see the correlation of each with the meditation activity. Thereupon I merged the dataframes into a unique dataset. I then splitted the data into features and target and I fit the model with data of the quartiles that presented higher correlation: Q2 and Q3. As my previous analysis showned, those quartiles were the most relevant as the person began entering a state of relaxation after 5 minutes into the meditation session. I fitted the data into different machine learning algohritms (Logistic Reggressor (LR), Random Forest Regressor (RFR) and K nearest neighbor (KNN)) and compared the accuracy scores of the predictions of each. I used crossed validation to avoid overfitting. After testing the models with different parameters, I choosed Logistic Regression as the best classifier, which an accuracy of 100% and a cross validation score average of 75%. I could consider myself satisfied with those results considering that I had only 16 columns to train and test the algohritm. More data is required to get better accuracy with machine learning.

## DATA ACQUISITION

```python
# Importing all dataframes and saving them into two different lists. 
# Associating each individual session with the name of the subject.
# Adding a column "Activity" to each dataframe. Associating the value 0 to normal state session and value 1 for meditation sessions.

import glob
path = "/Users/rebeccaestiarte/Desktop/IronHack/PROJECTS/Project-Week-8-Final-Project/your-project/Raw Data/" 
all_files = glob.glob(path + "/*.csv")

sessions_N = []
names_N = []
sessions_M = []
names_M = []

for filename in all_files:
    session = pd.read_csv(filename, index_col=None, header=0) # Reading file
    if "Normal" in filename: #Classifing as 0 if Normal Session
        session["Activity"] = 0 
        person = re.search("--\s{1}\w*\s{1}--", filename)[0].replace("--","").strip()
        names_N.append(person)
        sessions_N.append(session)
    elif "Meditative" in filename: #Classifing as 1 if Meditation Session
        session["Activity"] = 1 
        person = re.search("--\s{1}\w*\s{1}--", filename)[0].replace("--","").strip()
        names_M.append(person) 
        sessions_M.append(session)
```

## DATA WRANGLING

```python
# Cleaning all dataframes and saving into new list:
    # dropping columns out of score
    # removing NaN values
    # computing the average of all columns referring to same brainwave (e.g: ["Delta_TP9", "Delta_AF7", "Delta_AF8", "Delta_TP10"] -> Delta)

simplified_sessions_N = []
simplified_sessions_M = []

for session in sessions_N:
    session.drop(columns = ["AUX_RIGHT", "Accelerometer_X", "Accelerometer_Y", "Accelerometer_Z","Gyro_X","Gyro_Y", "Gyro_Z","HeadBandOn", "HSI_TP9","HSI_AF7","HSI_AF8","HSI_TP10","Battery","Elements"], inplace=True)
    session.dropna(inplace = True)
    session["Delta"] = session[["Delta_TP9", "Delta_AF7", "Delta_AF8", "Delta_TP10"]].mean(axis=1)
    session["Theta"] = session[['Theta_TP9', 'Theta_AF7', 'Theta_AF8', 'Theta_TP10']].mean(axis=1)
    session["Alpha"] = session[['Alpha_TP9','Alpha_AF7', 'Alpha_AF8', 'Alpha_TP10']].mean(axis=1)
    session["Beta"] = session[['Beta_TP9', 'Beta_AF7', 'Beta_AF8', 'Beta_TP10']].mean(axis=1)
    session["Gamma"] = session[['Gamma_TP9', 'Gamma_AF7', 'Gamma_AF8','Gamma_TP10']].mean(axis=1)
    
    simplified_session = session[["Delta", "Theta", "Alpha", "Beta", "Gamma", "Activity"]]
    simplified_sessions_N.append(simplified_session) 
    
for session in sessions_M:
    session.drop(columns = ["AUX_RIGHT", "Accelerometer_X", "Accelerometer_Y", "Accelerometer_Z","Gyro_X","Gyro_Y", "Gyro_Z","HeadBandOn", "HSI_TP9","HSI_AF7","HSI_AF8","HSI_TP10","Battery","Elements"], inplace=True)
    session.dropna(inplace = True)
    session["Delta"] = session[["Delta_TP9", "Delta_AF7", "Delta_AF8", "Delta_TP10"]].mean(axis=1)
    session["Theta"] = session[['Theta_TP9', 'Theta_AF7', 'Theta_AF8', 'Theta_TP10']].mean(axis=1)
    session["Alpha"] = session[['Alpha_TP9','Alpha_AF7', 'Alpha_AF8', 'Alpha_TP10']].mean(axis=1)
    session["Beta"] = session[['Beta_TP9', 'Beta_AF7', 'Beta_AF8', 'Beta_TP10']].mean(axis=1)
    session["Gamma"] = session[['Gamma_TP9', 'Gamma_AF7', 'Gamma_AF8','Gamma_TP10']].mean(axis=1)
    
    simplified_session = session[["TimeStamp", "Delta", "Theta", "Alpha", "Beta", "Gamma", "Activity"]]
    simplified_sessions_M.append(simplified_session)
```

Summarizing the data into two dataframe that contains, respectively, the mean of each brainwave frequencies at each datapoint(timestamp) of all meditation and normal state sessions. 

```python
df_concat = pd.concat((simplified_sessions_N[i] for i in range(len(simplified_sessions_N))))
by_row_index = df_concat.groupby(df_concat.index)
df_normal_means = by_row_index.mean().reset_index(drop=True)
df_normal_means
```

#### HYPOTESIS 1*: THE AVERAGE BRAIN FREQUENCIES AT MEDITATIVE STATE ARE CLOSER TO OPTIMAL RANGE(0)
*based on Observation 1

1. First, we want to reject the hypotesis of equality between brainwaves averages at normal vs meditative state:
2. Secondly, we will prove that the brainwaves average in meditation are lower than at normal state.
3. Finally, we will show how averages frequencies of the brainwaves at meditation are closer to the optimal range(0).

h0 : mu1 = mu2 
h1 : mu1 < mu2

where:
mu1 = average brain frequencies at normal state
mu2 = average brain frequencies at meditation state

Test Description: I will use the t-test to compare the mean of two given samples as the population parameters (mean and standard deviation) are not known. I proved above the normal distribution of the samples. I will use the independent samples t-test as I am comparing the mean for two independent groups.

Decision Rule: Reject H0 if the calculated p-value < .05.

```python
result = pd.concat([df_normal_means.drop(columns = "Activity"), df_meditation_means.drop(columns = "Activity")], axis=1).head(603)
columns = ['DeltaN', 'ThetaN',"AlphaN", "BetaN", "GammaN", 'DeltaM', 'ThetaM',"AlphaM", "BetaM", "GammaM"]
result.columns = columns
result['average_N'] = (result['DeltaN'] + result['ThetaN'] + result['AlphaN'] + result['BetaN'] + result['GammaN']) / 5
result['average_M'] = (result['DeltaM'] + result['ThetaM'] + result['AlphaM'] + result['BetaM'] + result['GammaM']) / 5
result = result[200:602] #We only take values from time200, as is when people get into the meditative state(5 min into the session, as showed in the graphs above)

X = sm.add_constant(result['average_N'])
model = sm.OLS(result.average_M, X)
results = model.fit()
print(results.summary())

mean_frequencies_norm = result['average_N'].mean()
mean_frequencies_med = result['average_M'].mean()

print ("We can observe from the model a low p-value (0.001), which is below our decision parameter  of 0.5 (see decision rule). We can therefore reject the hypotesis of equal averages with statistical significance.")
if mean_frequencies_norm > mean_frequencies_med:
    print("----------------------------------------")
    
    print("normal average:", mean_frequencies_norm)
    print("meditation average:", mean_frequencies_med)
    print("normal average < meditation average")
    print("We can clearly see that the average of brainwaves frequencies during meditation is closer to the optimal range of 0")
    print("----------------------------------------")
```

> Conclusion 1: We can conclude with statistical significance that the brainwaves during meditation have lower frequencies than at normal state. Also, they are closer to the optimal range of 0.

#### FEATURES CREATION

I am dividing each recorded session into four quartiles. I computed the mean of each brainwave frequency in each quartile. By concatenating all sessions with this processed information with 20 features (four quartiles per five brainwaves). The last column named "Activity" classifies the brainwave into meditation (1) and non-meditation (0). 

```python
all_sessions = simplified_sessions_N + simplified_sessions_M

def get_features(all_sessions, wave="Delta", df = pd.DataFrame()):
    firstQ = []
    secondQ = []
    thirdQ = []
    forthQ = []
    allQ = []

    for i in range(0,len(all_sessions)):
        total = len(all_sessions[i])
        mean_all = all_sessions[i][wave][:].mean()
        mean1 = all_sessions[i][wave][0: round(1/4 * total)].mean()
        mean2 = all_sessions[i][wave][round(1/4 * total): round(2/4 * total)].mean()
        mean3 = all_sessions[i][wave][round(2/4 * total): round(3/4 * total)].mean()
        mean4 = all_sessions[i][wave][round(3/4 * total): round(4/4 * total)].mean()
        firstQ.append(mean1)
        secondQ.append(mean2)
        thirdQ.append(mean3)
        forthQ.append(mean4)
        allQ.append(mean_all)

    #df = pd.DataFrame(columns=[wave+"_Q1", wave+"_Q2", wave+"_Q3", wave+"_Q4"])
    df[wave+"_Q1"] = firstQ
    df[wave+"_Q2"] = secondQ                  
    df[wave+"_Q3"] = thirdQ                   
    df[wave+"_Q4"] = forthQ  
    return df

cols = ['Alpha', 'Beta', 'Gamma', "Delta", "Theta"]
df = pd.DataFrame()

for wave in cols:
    df = get_features(all_sessions, wave, df)

df["Activity"] = 1
df.loc[0:len(simplified_sessions_N)-1]["Activity"] = 0
df
```
